A continuacion se presenta la implementacion de un algoritmo genetico para la minimizacion de una funcion de dos variables con restricciones.

* $max  f(x1,x2) = 13(x1-2)^2+9(x2-1)^2$

Restricciones:

* $ 0<=x_{1}<=4$
* $ 0<=x_{2}<=5$
* $ 0<x_{1}<4$
* $3x_{1}+4x_{2}-10$

# Estructura del algoritmo

Para la implementacion del algoritmo se siguio la estructura asociada a los algoritmos geneticos y para cada una de las etapas se implemento una funcion.

* generacion de la poblacion inicia **crearPoblacion()**
* decodificacion de la poblacion **decode()**
* seleccion de los individuos a cruzar **seleccion()**
* cruce de los individuos seleccionados **cruce()**
* mutacion **mutacion()**

Implementacion del programa principal:


``` java
clear
clc
NPOBLACION=100;
N1=5;
N2=5;
NUMITER=1000;
lamda=100;
resolucion=1/10;
nmutacion=3;
poblacion=crearPoblacion(NPOBLACION,N1,N2);
for i=1:NUMITER
    dpoblacion=decode(poblacion,N1,N2,resolucion);
    index=seleccion(dpoblacion,lamda);
    [dpoblacion , index, FA(dpoblacion(:,1),dpoblacion(:,2),lamda)]
    poblacion=cruce(poblacion,index,N1,N2,resolucion);
    poblacion=mutacion(poblacion,nmutacion);
end

dpoblacion=decode(poblacion,N1,N2,resolucion);
```


# Verificacion del algoritmo

Para realizar el proceso de verificacion del algoritmo implementado, este se comparo con la implementacion disponible en matlab de algoritmos geneticos. Esta implementacion solo resulve problemas de minimizacion por lo tanto se multiplico la funcion objetivo por -1.

Implementacion utilizando matlab las funciones **simple_fitness** y **simple_constraint**, corresponden a la funcion objetivo y las restricciones:


``` java
ObjectiveFunction = @simple_fitness;
nvars = 2;    % Number of variables
LB = [0 0];   % Lower bound
UB = [4 5];  % Upper bound
ConstraintFunction = @simple_constraint;
[x,fval] = ga(ObjectiveFunction,nvars,[],[],[],[],LB,UB, ...
    ConstraintFunction)
```

# Concluciones

Uno de los elementos que consideramos son definitivos en el funcionamiento del algoritmo son:  

**La codificacion**: despues de realizar pruebas de funcionamiento del algoritmo se decidio agrepar un factor de resolucion para la funcion de decodificacion este se fijo en 1/10. esto permite tener mayor resolucion en la busqueda de una solucion optima.

**El cruce** : La primer version de este modulo cuando el algoritmo estaba converguiedo a una solucion optima permitia que los hijos o la desendencia pudiera ser peor que los padres, lo que daba como resultado soluciones no optimas. por lo tanto se implemento dentro de el cruce, que los hijos tenian que ser mejor al menos que uno de los padres. esto permitio encontrar la solucion optima del problema. 

**El valor de lamda $\lambda$ **: En el caso en que las restricciones se tengan que cumplir de forma estriccta el valor de lamda juega un papel fundamental para que las solucion sea optima viable, en este caso quisimos hacer que las soluciones cumplieran de forma extricta con las restricciones por lo que para hallar el valor de lamda se realizaron diferentes pruebas hasta encontrar el valor optimo.
